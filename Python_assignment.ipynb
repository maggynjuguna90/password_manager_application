```python
file = "passwords.txt"
def password_manager():
    website = input("enter your website")
    email = input("enter your email")
    password = input("enter your password")
    try:
        with open("passwords.txt","a") as file:
            file.write(f"{website},{email},{password}\n")
        print("Password has been saved correctly ")
    except  Exception:
        print("Error saving password",Exception)
def retrieve_website():
    website  = input("Enter your website")
    try:
        with open ("passwords.txt","r") as file:
            found = False 
            for line in file:
                parts = line.strip().split(",", 2)
                if len(parts) != 3:
                    print("skipping badline",line)
                    continue
                w,e,p = parts
                if w.lower() == website.lower():
                    print(f'website: {w}')
                    print(f'email: {e}')
                    print(f'password: {p}')
                    found = True
                    break
            if not found:
                print("website not found")
    except FileNotFoundError:
        print("file not found")
def get_websites():
    try:
        with open ("passwords.txt","r") as file:
            websites = set()
            for line in file:
                parts = line.strip().split(",", 2)
                if len(parts) != 3:
                    print("skipping badline",line)
                    continue
                w,_,_ = parts
                websites.add(w)
            if websites:
                print("website saved correctly")
                for site in websites:
                    print("-",site)
            else:
                print("no website saved")
    except FileNotFoundError:
        print("file not found")


def users_menu():
    while True:
        print("\n ******welcome to password manager*****")
        print("1.password_manager")
        print("2.retrieve website")
        print("3.get websites")
        print("4.quit")
        choice = input("make a choice from above")
        if choice == "1":
         password_manager()
        elif choice == "2":
            retrieve_website()
        elif choice == "3":
         get_websites()
        elif choice =="4":
            print("Thank you for your time!")
            break
        else:
            print("make another choice")
users_menu()



            

```
